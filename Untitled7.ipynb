```python
# Sort DataFrame by 'Weight'
df_sorted_weight = df.sort_values(by='Weight', ascending=True)

# Get the bottom 20 cars by Weight (lightest)
df_bottom_20_weight = df_sorted_weight.head(20).copy()
df_bottom_20_weight['Weight Group'] = 'Bottom 20 Lightest'

# Get the top 20 cars by Weight (heaviest)
df_top_20_weight = df_sorted_weight.tail(20).copy()
df_top_20_weight['Weight Group'] = 'Top 20 Heaviest'

# Concatenate both DataFrames
df_combined_weight_groups = pd.concat([df_bottom_20_weight, df_top_20_weight])

print("Combined DataFrame for Top/Bottom 20 Weights (showing first 5 rows):")
display(df_combined_weight_groups.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
scaler_weight = MinMaxScaler()
scaler_mpg = MinMaxScaler()

# Normalize 'Weight' (lower weight is better, so invert the score)
df['Normalized_Weight'] = 1 - scaler_weight.fit_transform(df[['Weight']])

# Normalize 'MPG' (higher MPG is better)
df['Normalized_MPG'] = scaler_mpg.fit_transform(df[['MPG']])

# Create a combined score (higher is better for both 'inverted' Weight and 'direct' MPG)
df['Combined_Weight_MPG_Score'] = df['Normalized_Weight'] + df['Normalized_MPG']

# Sort by Combined_Weight_MPG_Score in descending order and get the top 10
top_10_cars_weight_mpg = df.sort_values(by='Combined_Weight_MPG_Score', ascending=False).head(10)

print("Top 10 Cars by Combined Weight (inverted) and MPG Score:")
display(top_10_cars_weight_mpg[['Car', 'Weight', 'MPG', 'Combined_Weight_MPG_Score']].reset_index(drop=True))
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Normalize 'Acceleration' and 'Horsepower'
df['Normalized_Acceleration'] = scaler.fit_transform(df[['Acceleration']])
df['Normalized_Horsepower'] = scaler.fit_transform(df[['Horsepower']])

# Create a combined score (higher is better for both)
df['Combined_Score'] = df['Normalized_Acceleration'] + df['Normalized_Horsepower']

# Sort by Combined_Score in descending order and get the top 10
top_10_cars = df.sort_values(by='Combined_Score', ascending=False).head(10)

print("Top 10 Cars by Combined Acceleration and Horsepower:")
display(top_10_cars[['Car', 'Acceleration', 'Horsepower', 'Combined_Score']].reset_index(drop=True))
```

```python
import numpy as np

# Get the indices of rows where 'MPG' is 0
zero_horsepower_indices = df[df['MPG'] == 0].index

print("Found 0 values in 'MPG' column. Please provide replacement values.")

for idx in zero_horsepower_indices:
    car_name = df.loc[idx, 'Car']
    print(f"\nCar: {car_name}")
    while True:
        try:
            new_hp = float(input(f"Enter new MPG for {car_name} (currently 0): "))
            if new_hp <= 0:
                print("MPG must be a positive number. Please try again.")
            else:
                df.loc[idx, 'MPG'] = new_hp
                break
        except ValueError:
            print("Invalid input. Please enter a numerical value for MPG.")

print("\nUpdated DataFrame after replacing 0 MPG values:")
display(df[df['MPG'] == 0])
print("Number of remaining zero MPG values:", (df['MPG'] == 0).sum())
```

```python
import plotly.express as px

# Calculate the average Horsepower for each country (Origin)
avg_horsepower_by_origin = df.groupby('Origin')['Horsepower'].mean().reset_index()

# Create a bar chart
fig_avg_horsepower_origin = px.bar(
    avg_horsepower_by_origin,
    x='Origin',
    y='Horsepower',
    title='Average Horsepower by Country of Origin',
    labels={'Origin': 'Country of Origin', 'Horsepower': 'Average Horsepower'},
    color='Origin' # Color bars by origin for better differentiation
)

fig_avg_horsepower_origin.show()
```
